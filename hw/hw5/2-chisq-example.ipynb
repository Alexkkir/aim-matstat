```r
dat <- iris

dat$size <- ifelse(dat$Sepal.Length < median(dat$Sepal.Length),
  "small", "big"
)
```

```r
head(dat)
```

```r
table(dat$Species, dat$size)
```

```r
library(ggplot2)

ggplot(dat) +
  aes(x = Species, fill = size) +
  geom_bar()
```

```r
ggplot(dat) +
  aes(x = Species, fill = size) +
  geom_bar(position = "dodge")
```

```r
test <- chisq.test(table(dat$Species, dat$size))
test
```
