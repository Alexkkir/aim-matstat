```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 10
beta = 3
x = np.linspace(0, 10, 100)
y = alpha * x + beta + np.random.normal(0, 1, 100)
plt.scatter(x, y)
```

```python
def estimations(x, y):
    x_ = np.mean(x)
    x2_ = np.mean(x**2)
    y_ = np.mean(y)
    xy_ = np.mean(x*y)
    beta = (xy_ - x_*y_) / (x2_ - x_**2)
    alpha = y_ - beta * x_
    return alpha, beta

alpha_, beta_ = estimations(x, y)
plt.scatter(x, alpha_ + x * beta_, color='red')
plt.scatter(x, y)
```

```python
def make_experiment(alpha, beta):
    x = np.linspace(0, 10, 100)
    y = alpha + x * beta + np.random.normal(0, 1, 100)
    return estimations(x, y)

experiments = [make_experiment(10, 3) for _ in range(1000)]
alphas, betas = zip(*experiments)
```

```python
plt.hist(alphas, bins=20)
```

```python
# check hypothesis that distribution of alpha is normal
from scipy.stats import shapiro
shapiro(alphas)
```

```python
# draw approximation of normal distribution for parameter alpha
from scipy.stats import norm
x = np.linspace(9, 11, 100)
plt.plot(x, norm.pdf(x, loc=np.mean(alphas), scale=np.std(alphas)))
plt.hist(alphas, bins=20, density=True)
```

```python
# draw approximation of normal distribution for parameter alpha
from scipy.stats import norm
x = np.linspace(2.9, 3.1, 100)
plt.plot(x, norm.pdf(x, loc=np.mean(betas), scale=np.std(betas)))
plt.hist(betas, bins=20, density=True)
```
