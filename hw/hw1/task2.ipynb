```r
n = 1000
theta=2
x = rexp(n, theta)
```

```r
t = seq(0, 10, 0.1)
f = exp(-theta * t)
```

```r
f_hat_func = function(t) {
    (1 - t / sum(x))^(n - 1) * (sum(x > t) > 0)
}
f_hat = sapply(t, f_hat_func)
plot(t, f, type="o", pch='', col='red')

lines(t, f_hat * 2, col='blue')
legend(1,  1, legend=c("f","f_hat"), col=c('red', 'blue'), pch=c("o","o"))
```
